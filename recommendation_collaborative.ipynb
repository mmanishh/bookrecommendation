```python
import numpy as np
import pandas as pd
from scipy.sparse import csr_matrix
import warnings
warnings.filterwarnings("ignore")
from sklearn.neighbors import NearestNeighbors

```

```python
books = pd.read_csv('data/books.csv', sep=';', error_bad_lines=False, encoding="latin-1")
books.columns = ['ISBN', 'bookTitle', 'bookAuthor', 'yearOfPublication', 'publisher', 'imageUrlS', 'imageUrlM', 'imageUrlL']
```

```python
users = pd.read_csv('data/users.csv', sep=';', error_bad_lines=False, encoding="latin-1")
users.columns = ['userID', 'Location', 'Age']
```

```python
ratings = pd.read_csv('data/ratings.csv', sep=';', error_bad_lines=False, encoding="latin-1")
ratings.columns = ['userID', 'ISBN', 'bookRating']
```

```python
combine_book_rating = pd.merge(ratings, books , on='ISBN')
```

```python
combine_book_rating = combine_book_rating[['userID','ISBN','bookRating','bookTitle','imageUrlS']]
```

```python
combine_book_rating.head()
```

```python
combine_book_rating.dropna(axis=0, subset=['bookTitle'],inplace=True)
```

```python
book_rating_count =combine_book_rating.groupby('bookTitle')['bookRating'].count().reset_index().rename(columns={'bookRating':'totalRatingCount'})[['bookTitle','totalRatingCount']]
```

```python
book_rating_count.sort_values(by='totalRatingCount',ascending=False)
```

```python
rating_with_total_ratingCount = combine_book_rating.merge(book_rating_count,left_on='bookTitle',right_on='bookTitle',how='outer')
```

```python
rating_with_total_ratingCount.head()
```

```python
pd.set_option('display.float.format',lambda x: '%.3f'%x)
```

```python
rating_with_total_ratingCount['totalRatingCount'].describe()
```

```python
popularity_threshold = 50
rating_popular_book = rating_with_total_ratingCount.query('totalRatingCount >=@popularity_threshold')
rating_popular_book.head()
```

```python
rating_popular_book.info()
```

```python
rating_popular_book_pivot = rating_popular_book.pivot_table(index='bookTitle',columns='userID',values='bookRating').fillna(0)
rating_popular_book_pivot_csr = csr_matrix(rating_popular_book_pivot.values)
```

```python
type(rating_popular_book_pivot_csr)
```

```python
model_knn = NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(rating_popular_book_pivot_csr)
```

```python
rating_popular_book_pivot_csr.shape
```
